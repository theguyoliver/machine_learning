## Feature Engineering

```python
credit_score.head(2)
```

### Encoding Categorical Data

```python
credit_score = credit_score.replace({
    'Gender': {'Female': 0, 'Male': 1},
    'Marital Status': {'Single': 0, 'Married': 1},
    'Home Ownership': {'Rented': 0, 'Owned': 1},
    'Credit Score': {'Low': 0, 'Average': 1, 'High': 2}
})
```

```python
credit_score['Education'] = credit_score['Education'].astype('category')
credit_score = pd.get_dummies(data=credit_score, columns=['Education'], prefix='Education')
credit_score.head(3)
```

```python
dummies = ["Education_Associate's Degree", "Education_Bachelor's Degree", "Education_Doctorate", "Education_High School Diploma", "Education_Master's Degree"]

for dummy in dummies:
    credit_score[dummy] = credit_score[dummy].astype(int)
credit_score.sample(3)
```

## Splitting Data into Training and Test Data

```python
x = credit_score.drop(columns=['Credit Score'], axis=1).values
y = credit_score['Credit Score'].values

x[:3], y[:3]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.15, random_state=42)
x.shape, x_train.shape, x_test.shape
```

## Modeling

```python
classifier = SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

## Model Evaluation

### On training data

```python
# Using Accuracy Score
train_data_preds = classifier.predict(x_train)
train_preds_acc = accuracy_score(y_train, train_data_preds)
train_preds_acc = round(train_preds_acc*100,2)
print(f"Model Accuracy on training data is {train_preds_acc}%")
```

```python
# Using Confusion Matrix

c_matrix = confusion_matrix(y_train, train_data_preds)
sns.heatmap(c_matrix, annot=True)
```

### On Test Data

```python
# Using Accuracy Score
test_data_preds = classifier.predict(x_test)
test_preds_acc = accuracy_score(y_test, test_data_preds)
test_preds_acc = round(test_preds_acc*100,2)
print(f"Model Accuracy on training data is {test_preds_acc}%")
```

```python
# Using Confusion Matrix

c_matrix_test = confusion_matrix(y_test, test_data_preds)
sns.heatmap(c_matrix_test, annot=True)
```
